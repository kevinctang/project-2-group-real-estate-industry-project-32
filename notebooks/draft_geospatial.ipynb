```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile using GeoPandas
shapefile_path = 'data/raw/SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp'
gdf = gpd.read_file(shapefile_path)

# Define the coordinates provided by the user
user_location = {'Latitude': -37.810272, 'Longitude': 144.962646}

# Create a GeoDataFrame for the user's location
user_gdf = gpd.GeoDataFrame([user_location], geometry=gpd.points_from_xy([user_location['Longitude']], [user_location['Latitude']]))

# Plot the shapefile with the user's location
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightgrey')
user_gdf.plot(ax=ax, color='red', markersize=50)

# Add a title and show the plot
plt.title('Location of User-Provided Coordinates on the Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import folium

# Define the coordinates
latitude = -37.810272
longitude = 144.962646

# Create a map centered around the user's coordinates
map_center = [latitude, longitude]
map_obj = folium.Map(location=map_center, zoom_start=13)

# Add a marker for the user's location
folium.Marker([latitude, longitude], popup="User Location").add_to(map_obj)

# Display the map
map_obj
```
